## 2.2 Build Data Loading Function 

```python
# Convert audio file to 16Hz
def load_wav_16k_mono(filename):
    
    # Load encobded wav file
    file_contents = tf.io.read_file(filename)
    
    # Decode wav (tensors by channels)
    wav, sample_rate = tf.audio.decode_wav(file_contents, desired_channels=1)
    
    # Remove trailing axis
    wav = tf.squeeze(wav, axis=-1)
    sample_rate = tf.cast(sample_rate, dtype=tf.int64)
    
    # 44100Hz --> 16000Hz - amplitude of the audio signal
    wav = tf_io.audio.resample(wav, rate_in=sample_rate, rate_out=16000)
    
    return wav
```

# 5. Build Preprocessing Function to Covert to Spectrogram 

## 5.1 Build Preprocessing Function 

```python
def preprocess(file_path, label):
    
    wav = load_wav_16k_mono(file_path)
    
    # Only get 3s
    wav = wav[:48000]
    
    # Pads
    zero_padding = tf.zeros([48000] - tf.shape(wav), dtype=tf.float32)
    wav = tf.concat([zero_padding, wav], 0)
    
    # Spectrogram
    spectrogram = tf.signal.stft(wav, frame_length=320, frame_step=32)
    spectrogram = tf.abs(spectrogram)
    spectrogram = tf.expand_dims(spectrogram, axis=2)
    
    return spectrogram, label
```

# 6. Create Traing and Testing Partitions 

## 6.1 Create a Tensorflow Data Pipeline 

```python
data = data.map(preprocess)
data = data.cache()
data = data.shuffle(buffer_size=1000)
data = data.batch(8)
data = data.prefetch(4)
```

## 6.2 Split into Training and Testing Partitions

```python
train_size = round(len(data) * .7)
test_size = len(data) - train_size

train_data = data.take(train_size)
test_data = data.skip(train_size).take(test_size)
```

# 7. Build Deep Learning Model 

## 7.1 Load Tensorflow Dependencies 

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, Dense, Flatten, MaxPooling2D
```

## 7.2 Build Sequential Model, Compile and View Summary 

```python
model = Sequential()
model.add(Conv2D(16, (3, 3), activation='relu', input_shape=(1491, 257, 1)))
model.add(MaxPooling2D(pool_size=(2, 2), strides=(2, 2)))
model.add(Conv2D(16, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2), strides=(2, 2)))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dense((1), activation='sigmoid'))
```

```python
model.compile('Adam', loss='BinaryCrossentropy', metrics=[tf.keras.metrics.Recall(), tf.keras.metrics.Precision()])
```

```python
model.summary()
```

# 9. Build Forest Parsing Functions 

## 9.1 Load up MP3s 

```python
def load_mp3_16k_mono(filename):
    
    res = tf_io.audio.AudioIOTensor(filename)
    
    # Convert to tensor and combine channels
    tensor = res.to_tensor()
    tensor = tf.math.reduce_sum(tensor, axis=1) / 2
    
    # Extract sample rate and cast
    sample_rate = res.rate
    sample_rate = tf.cast(sample_rate, dtype=tf.int64)
    
    # Resample to 16000Hz
    wav = tf_io.audio.resample(tensor, rate_in=sample_rate, rate_out=16000)
    
    return wav
```

## 9.2 Build Function to Convert Clips into Windowed Spectrograms

```python
def preprocess_mp3(samples, index):
    sample = samples[0]
    zero_padding = tf.zeros([48000] - tf.shape(sample), dtype=tf.float32)
    wav = tf.concat([zero_padding, sample], 0)
    
    spectrogram = tf.signal.stft(wav, frame_length=320, frame_step=32)
    spectrogram = tf.abs(spectrogram)
    spectrogram = tf.expand_dims(spectrogram, axis=2)
    
    return spectrogram
```

## 9.3 Convert Longer Clips into Windows and Make Predictions 

```python
audio_slices = tf.keras.utils.timeseries_dataset_from_array(wav, wav, sequence_length=48000, sequence_stride=48000, batch_size=1)
audio_slices = audio_slices.map(preprocess_mp3)
audio_slices = audio_slices.batch(32)
```

```python
y_hat = model.predict(audio_slices)
predictions = [1 if prediction > 0.9 else 0 for prediction in y_hat]

predictions
```

# 10. Make Predictions 

## 10.1 Loop over all recordings and make predictions 

```python
results = {}

for file in os.listdir(os.path.join('Data', 'Forest Recordings')):
    FILEPATH = os.path.join('Data', 'Forest Recordings', file)
    
    wav = load_mp3_16k_mono(FILEPATH)
    audio_slices = tf.keras.utils.timeseries_dataset_from_array(wav, wav, sequence_length=48000, sequence_stride=48000, batch_size=1)
    audio_slices = audio_slices.map(preprocess_mp3)
    audio_slices = audio_slices.batch(32)
    
    predictions = model.predict(audio_slices)
    
    results[file] = predictions
```

```python
len(results)
```

## 10.2 Convert Predictions to Classes 

```python
class_preds = {}
for file, logits in results.items():
    class_preds[file] = [1 if prediction > 0.9 else 0 for prediction in logits]
    
class_preds
```

## 10.3 Group Consecutive Detections

```python
post_processed = {}
for file, scores in class_preds.items():
    post_processed[file] = tf.math.reduce_sum([key for key, group in groupby(scores)]).numpy()
    
post_processed
```
